```python
# usual step to import the library
import seaborn as sns
```

```python
%matplotlib inline # seaborn is built on top of matplotlib and also to see our visua;ization since we are inside our notebook.
```

```python
tips = sns.load_dataset('tips') # this is seaborn's built in datasets. 
```

```python
sns.set_style("darkgrid")
tips.head(5) # this is tips data, basically it is a restaurant data of poeple leaving the tips after their meal
            # with variables like gender, total bill time(whether lunch or dinner etc), which day of the week 
            # whether smoker or non-smoker
```

```python
# first plot 
# dist plot - It is basically shows us the distribution of univariate set of observations(single variable).
# you need to pass single column (the variable (single)) and it wil show the distribution

#sns.distplot(tips['total_bill'])
# the line is kde(kernel density estimation) we can also remove it we can also set the bins because with these default bins the distribution is not clear
sns.distplot(tips['total_bill'],kde = False, bins = 35)

# this is basically a histogram of total bill
# A Histogram visualises the distribution of data over a continuous interval or certain time period. 
#Each bar in a histogram represents the tabulated frequency at each interval/bin.
# Histograms help give an estimate as to where values are concentrated,
#what the extremes are and whether there are any gaps or unusual values. 
#They are also useful for giving a rough view of the probability distribution.
#A histogram is a graphical representation that organizes a group of data points into user-specified ranges. 
#It is similar in appearance to a bar graph. The histogram condenses a data series into an easily interpreted visual 
#by taking many data points and grouping them into logical ranges or bins.

# In this is graph on the y-axis you see the frequencies 
# on the x- axis you see the bill value range 
# we can see the most the bills were (you can see most of the bill values are between 10-20)
# if you increase the bins it will try to plot each and every single instance for every single price point.
```

```python
# Joint plots
# joint plots basically allows you to plot or compare (match up) for bivariate data. we can plot distributions for 
 #two variables 
# (continuos)
# there is a kind parameter that allows us how we want to combine our data
# arguments x for vraible1 and y for variable2 and data for data.
sns.jointplot(x='total_bill',y='tip', data =tips, kind ='hex')

# we can chage the kind argument and play with it by changing the type reg is one kde , hex, scatter is default and 
# some others


```

```python
# Some notes about the plot above

# darker hexagon mean more values around that region (more values around that range). semi dark means relative less value 

# let us draw a plot with kde
sns.jointplot(x='total_bill',y='tip', data =tips, kind ='kde')
```

```python
# this is a kde plot - it is quite evident that the darker in the plot suggest maximum values(instances), 
# values decreases as it fades. 
# on the other plot we see two lumps which represent the more values in that range 
```

```python
# this reg plot will draw a regressionl line in the scatter plot 

#sns.jointplot(x='total_bill',y='tip', data =tips, kind ='reg')

sns.jointplot(x='total_bill',y='tip', data =tips, kind ='scatter') # this is with kind = scatter
```

```python
# pairplot
# they from pairwise relationships across the entire dataset at least for numerical columns.
# we just need to pass the dataset name
sns.pairplot(tips)
```

```python
# there is also a pair plot which does every possible combinations of the numerical columns( continuous variables) in a
# dataframe.
#sns.pairplot(tips) 
# It also has an important feature hue which essentially allows us to include a categorical variable with these plots 
#sns.pairplot(tips, hue = 'sex') # we can understand the males and females with the legend 
# we can also change the color with palette argument
sns.pairplot(tips, hue = 'sex', palette ='Reds_r')
# one hack if you want to know what are the values that are accepted in the palette just type 'orange' it will show an error
# in the error it will show all the colors which are accepted in the palette.
```

```python
# RUG PLOT
# it is a very simple plot which draws(plots) every single point of a single variable(univariate) on the distribution line 
sns.rugplot(tips['total_bill'], height = 0.2)
# here also we can pass only one variable (a univariate plot)
# it has height argument through which we can adjust the height of the lines
#(these lines denotes every single instance of the variable total_bill on 'x' axis)
```

```python
# there is kernel density plot which are build off of these rugs plot
# it is basically summation of all the normal distribution plots build off of rug plots
```

```python
## Categorical Plots
# barplot with mean
sns.barplot(x='sex',y='total_bill', data = tips)

# this is bar plot with mean of the categorical data on y axis and 
# x-axis denotes categories
# we can change the mean to count, sum and other functions in the bar plot
# there is an estimator function which lets you change the function by default it mean 
# essentially estimatror can take all the aggregate function.
```

```python
tips.groupby('sex').mean()['total_bill']
```

```python
# change the estimator
import numpy as np
sns.barplot(y='sex',x='total_bill', data = tips, estimator =np.std)
# we can see the estimator changed to standard deviation 
# standard deviation of male is higher than the the female but but number of males are also higher than the females
```

```python
# there is also a count plot which is used only to see the number of occurrences
sns.countplot(x='sex', data = tips)
```
