```python
import seaborn as sns
% matplotlib inline
tips = sns.load_dataset('tips')
tips.head(5)

# in this we see plots of categorical variables like sex, day, time etc and we will reference them with numerical variable.
```

```python
# first is barplot
# Essentially barplot allows you to aggregate categorical variable based off of some function by default it is mean.
# Basically it is a visualization of a group by action.
sns.barplot(x='sex',y='total_bill', data = tips)

# this is bar plot with mean of the categorical data on y axis and 
# x-axis denotes categories
# we can change the mean to count, sum and other functions in the bar plot
# there is an estimator function which lets you change the function by default it mean 
# essentially estimatror can take all the aggregate function.
```

```python
# observation of the above plot
# y-axis represents total _avg bills
# avg bill of male is slightly higher than the female

# let us verify this by using normal group by function

tips.groupby('sex').mean()['total_bill']
```

```python
# count plot specifically plots the number of occurences of a categorical value 
#  (it is very to bar plot) here the estimator is the count
# we will not set the x value

sns.countplot(x='sex', data = tips)

# this is 
```

```python
# Box plots 
# this is also known as box and whisker plots and it shows the ditribution of quantitative data in a way 
# that facilitates comparison between variables
sns.boxplot(x='day', y= 'total_bill', data = tips)

# here we can see the distribution per day.
```

```python
# we can also add an argument to our box plot which allow these boxplots to split up by argument "hue" which will aid us.
# we can add another categorical column with our hue argument in this case it is smoler which will split up the box plot
# it is split per day and also split by smoker status. we can see now day by day bill and also whether a person is a smoker.

sns.boxplot(x='day', y= 'total_bill', data = tips, hue = "smoker")
```

```python
# violin plots
#sns.violinplot(x= 'day', y= 'total_bill', data = tips)

# violin plots are basically boxplot and histogram 
# thick black area is a box plot narrowing down to whiskers at the top and bottom 
# on the sides you see a normal distribution (probability density function)
# this also has hue argument. It also has a split argument which makes hue's variable more evident and easy to read.
# split argument avoids unecessary plots.



```

```python
# example
sns.violinplot(x='day', y = 'total_bill', data = tips, hue = 'sex')
```

```python

# now split = True
sns.violinplot(x='day', y = 'total_bill', data = tips, hue = 'sex' , split = True)
```

```python
# Stripplot

sns.stripplot(x='day', y = 'total_bill', data = tips)

# drawback we wont be bale to know how many points are stacked on top of each other.
# though we can add jitter argument to add some randomness in points to get a good view
```

```python
# Example with jitter
sns.stripplot(x='day', y = 'total_bill', data = tips, jitter = True)
# we can also use split and hue on this also
```

```python
# swarmplots
# seaborn also gives us swarmplot which is like a violin in striplot 
# they are visually ok but has drawback with large amount of data where it is difficut to interpret how is the distribution 
# and how many points are actually there

sns.swarmplot(x='day', y = 'total_bill', data = tips)

# you can combine violin ploy with these 
# there are similar arguments in this also.
```

```python
# Factorplot
# Factorplot is the most general form of all these plots with its kind argument we can plot all these plot.
sns.factorplot(x='day', y = 'total_bill', data = tips)
sns.factorplot(x='day', y = 'total_bill', data = tips, kind = 'bar')
```
