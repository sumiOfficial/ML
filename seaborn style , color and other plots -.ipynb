```python
# Regression plots
# lmplots allows you to display linear models with seaborn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x= 'total_bill', y='tip', data = tips ) 
# we can change add hue and also change the style and color
```

```python
# hue added

sns.lmplot(x= 'total_bill', y='tip', data = tips, hue = 'sex') 
```

```python
# instead of doing with the hue we can separate with col arguemnt
sns.lmplot(x= 'total_bill', y='tip', data = tips, col = 'sex') 
# here we get two separate plots separated by sex
```

```python
# we can also add row argument with another categorical variable
sns.lmplot(x= 'total_bill', y='tip', data = tips, col = 'sex', row = 'time') 
```

```python
# we can also change the size of the circles in the plot and add something like a triangle etc 
# seaborn is library built on top of matplotlib 

sns.lmplot(x= 'total_bill', y='tip', data = tips, hue = 'sex' , markers = ['o', 'v'],
          scatter_kws={'s': 80}) 
# markers are from matplotlib they specify the dots and triangle. since we used hue we need two markers for our plot.
# scatter_kws is a dictionary which allows us to increase the size of these dots and traingle
```
