# Object Oreinted Method

In object orinted method we use an empty canvas to draw our figures and it is mainly useful when we want multiple figures in the 
the same canvas.
The idea behind object Oriented method is to create figure objects and then just call methods or attributes off of that object.

```python
# let's create an empty canvas
canv = plt.figure()

# Now let us add axes to the figure
# Add axes also means putting a certain percentage criteria on the axes 
#covering left, bottom, width, height ranging from (range 0 to 1)
axes = canv.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel("X Axis")
axes.set_ylabel("Y Axis")
axes.set_title(" x vs y^2")
```

```python
# let us see the real benefit of oo method
```

```python
canv1=plt.figure()
# adding axes
axes1 = canv1.add_axes([0.1,0.1,0.9,0.9])
axes1.plot(x,y,'b')
axes2 = canv1.add_axes([0.2,0.5,0.4,0.4])
axes2.plot(y,x,'g')
```

```python
# Horizontal bar plots
# Example data
plt.rcdefaults()
fig, ax = plt.subplots()

people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
ax.bar(y_pos, performance, xerr=error,align='center',
        color='green', ecolor='black')
```

```python
ax.barh(y_pos, performance, xerr=error, align='center',
        color='green', ecolor='black')
```
