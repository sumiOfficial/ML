```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

# let us see first the simple functional method 
# Functional

```python
plt.plot(x,y)
# this is a very simple plot to show how plots in jupyter(python) works 
# we can also change color of the plot put labels and titles 
# let us see how we can do that 
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("First Plot x to Y^2")
```

```python
# let us see now if we can create two or more plots on the canvas 
# for that python has "Subplots" methods.
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)
```
